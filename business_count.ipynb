```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# loading and compiling each dataset associated to each year and creating column for year
year = ['2011','2012','2013','2014','2015','2016','2017','2018']
data_ = pd.DataFrame()

zbp = pd.read_csv("data/zbp11totals.csv")
zbp['year'] = 2011
print(zbp.shape)

p1 = pd.read_csv("data/zbp12totals.csv")
p1['year'] = 2012
zbp = zbp.append(p1)

print(zbp.shape)

p2 = pd.read_csv("data/zbp13totals.csv")
p2['year'] = 2013
zbp = zbp.append(p2)
print(zbp.shape)

p3 = pd.read_csv("data/zbp14totals.csv")
p3['year'] = 2014
zbp = zbp.append(p3)
print(zbp.shape)

p4 = pd.read_csv("data/zbp15totals.csv")
p4['year'] = 2015
zbp = zbp.append(p4)
print(zbp.shape)

p5 = pd.read_csv("data/zbp16totals.csv")
p5['year'] = 2016
zbp = zbp.append(p5)
print(zbp.shape)

p6 = pd.read_csv("data/zbp17totals.csv")
p6['year'] = 2017
zbp = zbp.append(p6)
print(zbp.shape)

p7 = pd.read_csv("data/zbp18totals.csv")
p7['year'] = 2018
zbp = zbp.append(p7)
print(zbp.shape)
```

```python
# creating copy to refer back to
zbp_orig = zbp.copy()

# data dictionary link:
# https://www2.census.gov/programs-surveys/cbp/technical-documentation/records-layouts/2017_record_layouts/zip_code_totals_record_layout_2017.txt
```

```python
# null values
zbp.isnull().sum()
```

```python
# adding leading zeros to all zipcodes with less than 5 digits
print(type(zbp['zip'].iloc[0]))
zbp['zip'] = zbp['zip'].astype(str).str.zfill(5)
```

```python
# dropping columns we don't need
zbp.drop(['emp_nf', 'qp1_nf', 'ap_nf', 'name', 'cty_name'], axis=1, inplace=True)
```

```python
emp_dict = {'A':10, 'B':60, 'C':175, 'E':375, 
            'F':750, 'G':1750, 'H':3750, 'I':7500, 
            'J': 17500, 'K':37500, 'L': 75000, 
            'M':100000, 'S':np.NaN}
emp_list = list(emp_dict.keys())
```

```python
for i in range(zbp.shape[0]):
    if zbp['empflag'].iloc[i] in emp_list:
        zbp['emp'].iloc[i] = emp_dict[zbp['empflag'].iloc[i]]
        
zbp
```

```python
# only interested in the number of business establishments per zip per year
zbp.drop(['empflag', 'emp', 'qp1', 'ap', 'city', 'stabbr'], axis=1, inplace=True)
```

```python
# convert to csv
zbp.to_csv('bus_count.csv', index=False)
```
